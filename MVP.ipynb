Vacation Recommendations


The goal of this project is to create a recommendation system to make a city recommendation for a vacation based on the seasonality of the city's attractions.

Data

I webscraped attraction reviews and ratings from TripAdvisor. I focus only Top 10 'Things to Do' in cities of Austin, Chicago, Seattle, and New York. Hotels, restaurants, and landmarks are not included. The data consists of over 24,000 user reviews. 

Work Completed

I preprocessed Austin and Chicago data by removing digits and punctuation.
To analyze the seasonality of users' reviews of each attraction I used Vader sentiment analyzer. I also aggregated reviews of top 10 attractions of each city and computed the average sentiment score per city per month.

To build a topic model, I created a baseline model using CountVectorizer with LSA topic modeler and a small number of components (n=6). I also created a separate baseline model for negative reviews (with less than 4 stars) to analyze the reasons underlying the seasonality of the cities' popularity.

Recent Findings

I found that the users' perception of the attractions in Austin and Chicago are indeed seasonal. 
For example, the next plot shows that Congress Avenue Bridge in Austin is quite popular in late winter when it is not too hot to explore trails under the bridge and during the bat season (March-September) when Mexican bats live in it and make their impressive nightly departure before the sunset. However, it is not as popular after the bats migrate back to Mexico and in late fall when it gets rainy and cold.

The next plot shows that the vader score of the Millenium park in Chicago is quite high during the warm months and, surprisingly, in winter too. I explored the reviews of the park in December and January to understand the reason for these high sentiment scores and learned that visitors greatly enjoy the ice rink installed in the park during the winter months.
